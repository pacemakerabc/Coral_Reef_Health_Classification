```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/coral_cleaned.csv")
df.head()
```

```python
df.info()
```

```python
numeric_columns = ["Temperature Stress Anomalies Frequency Max", "Temperature Stress Anomalies Frequency Mean", "Temperature Stress Anomalies Degree Heating Weeks", "Temperature Stress Anomalies Degree Heating Weeks Maximum", "Temperature Stress Anomalies Degree Heating Weeks Mean"]
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
for i in numeric_columns:
  df.dropna(subset=[i], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
for i in ["Ocean_Name", "Country_Name", "Bleaching Comments"]:
  df[i] = Le.fit_transform(df[i])
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:, 0:9]
Y = df["Bleaching Comments"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Model Accuracy:", accuracy_score(Y_test, Y_pred))
print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
print("Classification Report:\n", classification_report(Y_test, Y_pred))
```
